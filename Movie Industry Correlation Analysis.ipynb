```python
# Let's Import the libraries

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None



# Now we need to read in the data
df = pd.read_csv(r'C:\Users\Swapnil\Downloads\movies\movies.csv')
```

```python
#Lets look at the data

df.head()
```

```python
# Lets Explore any missing data
# Let's loop through the data and see if there is anything missing

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
# Let's remove the rows with missing values

df = df.dropna()
```

```python
df.head()
```

```python
# Data types in our columns

df.dtypes
```

```python
# Now lets convert the data type to int

df['budget'] = df['budget'].astype('int64')

df['gross'] = df['gross'].astype('int64')
```

```python
# there is an inconsistancy in column YEAR and RELEASED. Let's resolve the issue
# LEts create a new column and name it - YEARCORRECT

def extract_year(released_str):
    if isinstance(released_str, str):
        for part in released_str.split(' '):
            if part.isdigit() and len(part) == 4:
                return part
    return None  # return None if the entry is not a string or doesn't contain a year

df['yearcorrect'] = df['released'].apply(extract_year)
```

```python
df.head()
```

```python
df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized.head()
```

```python
correlation_matrix = df_numerized.corr(method = 'pearson')

sns.heatmap(correlation_matrix, annot = True)

plt.title('Correlation Matric for Numerical Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

plt.show()
```

```python
df_numerized.corr()
```

```python
correlation_mat = df_numerized.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5 ]

high_corr
```
